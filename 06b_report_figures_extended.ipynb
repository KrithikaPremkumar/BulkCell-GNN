# 06b — Extended 300-DPI report figures

Re-export six key figures into `DATA_DIR/figures_extended`.


## Setup


```python
from pathlib import Path
import os, sys, json, time, random
import numpy as np
import pandas as pd

USE_DRIVE = True
SEED = 42

if USE_DRIVE:
    from google.colab import drive
    drive.mount('/content/drive')

DATA_DIR = Path('/content/drive/MyDrive/BulkCellGNN_data')
DATA_DIR.mkdir(parents=True, exist_ok=True)
sys.path.insert(0, '/content/drive/MyDrive/BulkCellGNN_data')

random.seed(SEED)
os.environ['PYTHONHASHSEED'] = str(SEED)
np.random.seed(SEED)

print('DATA_DIR =', DATA_DIR)

import torch
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print('device =', device)

FIG_DIR = DATA_DIR / 'figures_extended'
FIG_DIR.mkdir(parents=True, exist_ok=True)
print('FIG_DIR =', FIG_DIR)
```

## Figure map


```python
figure_map = {
    'cmp_section2_roc.png': 'fig1_three_model_roc_300dpi.png',
    'cmp_section3_gamma_heatmaps.png': 'fig2_three_model_gamma_heatmaps_300dpi.png',
    'cmp_section5_alpha_convergence.png': 'fig3_shilu_parameter_evolution_300dpi.png',
    'best_model_bulk_umap.png': 'fig4_best_model_bulk_umap_300dpi.png',
    'best_model_cell_umap.png': 'fig5_best_model_cell_umap_300dpi.png',
    'cmp_section2_metrics_bar.png': 'fig6_auc_summary_bar_300dpi.png',
}
```

## Export 300-DPI figures


```python
for src_name, dst_name in figure_map.items():
    src = DATA_DIR / src_name
    if not src.exists():
        print('MISSING source:', src_name)
        continue
    img = mpimg.imread(src)
    h, w = img.shape[:2]
    fig = plt.figure(figsize=(w / 300, h / 300), dpi=300)
    ax = fig.add_axes([0, 0, 1, 1])
    ax.imshow(img)
    ax.axis('off')
    out = FIG_DIR / dst_name
    fig.savefig(out, dpi=300, facecolor='white', bbox_inches='tight', pad_inches=0)
    plt.close(fig)
    print('Saved', out.name)
```

## Final verification


```python
expected = ['fig1_three_model_roc_300dpi.png','fig2_three_model_gamma_heatmaps_300dpi.png','fig3_shilu_parameter_evolution_300dpi.png','fig4_best_model_bulk_umap_300dpi.png','fig5_best_model_cell_umap_300dpi.png','fig6_auc_summary_bar_300dpi.png']
missing = []
for name in expected:
    p = FIG_DIR / name
    if p.exists():
        print('OK ', name, round(p.stat().st_size / 1024, 1), 'KB')
    else:
        print('MISSING', name)
        missing.append(name)
if missing:
    raise FileNotFoundError('Missing figure exports: ' + ', '.join(missing))
print('Notebook 06b complete')
```
